```python
from pathlib import Path
import importlib.util
import sys

# Load the sibling module by file path so the notebook works even when VS Code
# starts the kernel from a different working directory.
TOOL_PATH = Path(r"C:\Users\USER\OneDrive\Desktop\zip\zip\web_search\web_search_tool.py")

spec = importlib.util.spec_from_file_location("web_search_tool", TOOL_PATH)
if spec is None or spec.loader is None:
    raise ImportError(f"Could not load web_search_tool from {TOOL_PATH}")

web_search_tool = importlib.util.module_from_spec(spec)
sys.modules["web_search_tool"] = web_search_tool
spec.loader.exec_module(web_search_tool)

web_search = web_search_tool.web_search
TOOL_DESCRIPTION = web_search_tool.TOOL_DESCRIPTION

print("web_search() ready.")
print(TOOL_DESCRIPTION)
```
