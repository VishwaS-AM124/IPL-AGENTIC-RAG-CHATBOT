## 1. Load Datasets

```python
from pathlib import Path
import importlib.util
import sys

# Load the sibling module by file path so VS Code/Pylance does not flag the
# notebook import even when the notebook working directory is different.
TOOL_PATH = Path(r"C:\Users\USER\OneDrive\Desktop\zip\zip\query_data\query_data_tool.py")

spec = importlib.util.spec_from_file_location("query_data_tool", TOOL_PATH)
if spec is None or spec.loader is None:
    raise ImportError(f"Could not load query_data_tool from {TOOL_PATH}")

query_data_tool = importlib.util.module_from_spec(spec)
sys.modules["query_data_tool"] = query_data_tool
spec.loader.exec_module(query_data_tool)

query_data = query_data_tool.query_data
matches_df = query_data_tool.matches_df
ball_df = query_data_tool.ball_df
players_df = query_data_tool.players_df
ALL_TEAMS = query_data_tool.ALL_TEAMS

print("matches_df :", matches_df.shape, "|", [c for c in matches_df.columns if c != "_source_row"])
print("ball_df    :", ball_df.shape, "|", [c for c in ball_df.columns if c != "_source_row"])
print("players_df :", players_df.shape, "|", [c for c in players_df.columns if c != "_source_row"])
print("Teams      :", ALL_TEAMS)
```

## 3. Test Queries

```python
import pandas as pd


def run(q):
    print(f"\nQUERY: {q}")
    res = query_data(q)
    if res.get("error"):
        print("ERROR:", res["error"])
        return res

    if isinstance(res["result"], list):
        print(f"row_count: {res['row_count']}")
        print(f"columns  : {res['columns']}")
        print(pd.DataFrame(res["result"]).head(10).to_string(index=False))
    else:
        print(f"result   : {res['result']}")

    print("source  :", res.get("source"))
    return res
```

```python
run("Who won final match in 2023?")
run("Match winners between Chennai Super Kings and Mumbai Indians")
run("run of kohli 2023")
run("runs of Virat Kohli in 2023")
run("How many runs did Virat Kohli score in 2023?")
run("strike rate of Virat Kohli in 2023")
run("How many wickets did Mohammed Shami take in 2023?")
run("top 5 run scorers in 2024")
run("How many matches did RCB win in 2024?")
```
