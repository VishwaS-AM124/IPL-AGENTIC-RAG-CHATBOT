# IPL Search Docs Tool

Reusable document-search tool for the Agentic RAG assignment. The implementation lives in `search_docs_tool.py`; this notebook is only for testing and demonstration.


```python
from pathlib import Path
import importlib.util
import sys

TOOL_PATH = Path(r"C:\Users\USER\OneDrive\Desktop\zip\zip\search_docs\search_docs_tool.py")

spec = importlib.util.spec_from_file_location("search_docs_tool", TOOL_PATH)
if spec is None or spec.loader is None:
    raise ImportError(f"Could not load search_docs_tool from {TOOL_PATH}")

search_docs_tool = importlib.util.module_from_spec(spec)
sys.modules["search_docs_tool"] = search_docs_tool
spec.loader.exec_module(search_docs_tool)

search_docs = search_docs_tool.search_docs
parse_query = search_docs_tool.parse_query
TOOL_DESCRIPTION = search_docs_tool.TOOL_DESCRIPTION

print("search_docs() ready.")
print(TOOL_DESCRIPTION)
```

```python
def show(query, top_k=3):
    result = search_docs(query, top_k=top_k)
    print(f"\nQUERY: {query!r}")
    print("parsed:", parse_query(query))
    if result.get("error"):
        print("ERROR:", result["error"])
        return result
    for item in result["results"]:
        print(f"\n#{item['rank']} score={item['score']} {item['citation']} reason={item['reason']}")
        print(item["text"][:700])
    return result
```

```python
show("Which team won Match 1 of IPL 2023?")
show("Who won Match 1 of IPL 2024?")
show("Who won the IPL 2024 final?")
show("Virat Kohli brilliant batting performance")
show("Premier League football")
show("")
```
